```python
import random

# Define arithmetic operations
operations = ['+', '-', '*', '/']

# Function to compute the result of an arithmetic operation
def compute_result(num1, num2, operator):
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        if num2 == 0:  # Avoid division by zero
            return None
        return num1 // num2  # Integer division
    return None

# Function to generate a random quadruple
def generate_quadruple():
    while True:
        num1 = random.randint(0, 99)
        num2 = random.randint(0, 99)
        operator = random.choice(operations)
        result = compute_result(num1, num2, operator)
        
        # Ensure the result is valid and within the range 0-99
        if result is not None and 0 <= result <= 99:
            return (num1, operator, num2, result)

# Function to create a set of unique quadruples
def create_dataset(size, exclude_set=None):
    dataset = set()
    
    while len(dataset) < size:
        quadruple = generate_quadruple()
        if exclude_set is None or quadruple not in exclude_set:
            dataset.add(quadruple)
    
    return dataset

# Create the test set of 100 quadruples
test_set = create_dataset(100)

# Create the training set of 10,000 quadruples, excluding the test set
training_set = create_dataset(10000, exclude_set=test_set)

# Convert the sets to lists for easier handling (optional)
test_set = list(test_set)
training_set = list(training_set)

# Display the first 5 quadruples from each set as a sample
print("Test Set (sample):", test_set[:5])
print("Training Set (sample):", training_set[:5])

# Save the datasets to a file (optional)
def save_dataset(dataset, filename):
    with open(filename, 'w') as f:
        for quadruple in dataset:
            f.write(f"{quadruple[0]} {quadruple[1]} {quadruple[2]} = {quadruple[3]}\n")

# Save test and training sets
save_dataset(test_set, './test_set.txt')
save_dataset(training_set, './training_set.txt')
```
